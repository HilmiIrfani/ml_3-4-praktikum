**Finding Outliers with the 3-Sigma Rule**

```python
mean = df_noid['wins'].mean()
stdev = df_noid['wins'].std()

iso = (df_noid['wins'] < mean - 3 * stdev) | (df_noid['wins'] > mean + 3 * stdev)

df1 = df_noid[['team', 'wins']].assign(is_outlier=iso)
print(df1)
```
